```python
import pandas as pd 
import json
import csv
from pandas import json_normalize
from flatten_json import flatten
```

```python
f = open('Datasets/balances.json') 
dfbalance = json.load(f) 
f.close()
```

```python
dfbalance
```

```python
df = pd.DataFrame([key for key in dfbalance.keys()], columns=['id_cliente'])
```

```python
df.head(5)
```

```python
df['prueba'] = [value[::] for value in dfbalance.values()]
```

```python
df_balance1 = df.explode('prueba')
```

```python
df_balance1.head(5)
```

```python
df_balance2 = pd.json_normalize(df_balance1['prueba'][::], errors='ignore',)
```

```python
df_balance2.head(5)
```

```python
df_balance2.shape
```

```python
df_id_cliente = df_balance1['id_cliente']
```

```python
df_id_cliente = pd.DataFrame(df_id_cliente)
```

```python
df_id_cliente.head(5)
```

```python
df_id_cliente.shape
```

```python
df_balance3 = pd.merge(df_id_cliente, df_balance2, left_index = True, right_index=True
) 
```

```python
df_balance3.tail(5)
```

```python
df_balance3.to_csv('df_balance.csv', index = False)
```
